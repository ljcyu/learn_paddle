```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from pathlib import Path

for i, image in enumerate(Path('data/zebra crossing').glob('*.png')):
    if i%100==0:
       fig = plt.figure(figsize=(12,18))
       gs = plt.GridSpec(10, 10)
       gs.update(wspace=0.05, hspace=0.3)

    ax = plt.subplot(gs[i%100])
    plt.axis('off')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_aspect('equal')
    plt.title(image.stem)
    #plt.imshow(np.squeeze(img).astype('uint8'), cmap='gray')
    img=Image.open(image)
    plt.imshow(img)

plt.show()
```
