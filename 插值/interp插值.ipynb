```python
# 根据x、y的关系进行插值
import numpy as np
from numpy import interp
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

xvals = np.linspace(0, 2*np.pi, 50)
yinterp = np.interp(xvals, x, y)

import matplotlib.pyplot as plt
plt.plot(x, y, 'o')
#[<matplotlib.lines.Line2D object at 0x...>]
plt.plot(xvals, yinterp, '-x')
#[<matplotlib.lines.Line2D object at 0x...>]
plt.show()
```
