```python
#命令行可以运行，无法在jupyter notebook运行
import paddle
import numpy as np
import paddle.nn as nn
import paddle.nn.functional as F
net=paddle.nn.Linear(1,1)
optimizer=paddle.optimizer.SGD(learning_rate=0.1,parameters=net.parameters())

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype='float32').reshape(-1,1)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype='float32').reshape(-1,1)

    
x=paddle.to_tensor(xs)
y=paddle.to_tensor(ys)
print(x.shape,y.shape)

for i in range(100):
    y_hat=net(x)
    
    loss = F.square_error_cost(y_hat, label=y)                   
    avg_loss=paddle.mean(loss)
    print(avg_loss.numpy())
    avg_loss.backward()
    optimizer.step()
    optimizer.clear_grad()

print(net.weight,net.bias)        

net.eval()
data=np.array([10,3,4],dtype='float32').reshape(-1,1)
data=paddle.to_tensor(data)
print(net(data))
```
