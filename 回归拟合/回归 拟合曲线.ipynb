```python

x=np.array([1,2,3,4]).reshape((4,1))
y=np.array([2,4,6,8]).reshape((4,1))
x
y
```

```python
a=0
b=0
n=len(x)
lr=0.001
for i in range(1000):
    y_predict=a*x+b
    #print(y_predict)
    loss=0        
    loss=(y_predict-y)*(y_predict-y)
    #print('loss=',loss)
    avg_loss=np.sum(loss,axis=0)/(2*n)
    print('avg_loss=',avg_loss)
    da=np.sum((y_predict-y)*x,axis=0)/n
    db=np.sum(y_predict-y,axis=0)/n
    #print('da=',da)
    a=a-lr*da
    b=b-lr*db
    print('{}x+{}'.format(a,b))
    if avg_loss<1e-5:
        print('准确度已经很高，break')
        break
            
```
