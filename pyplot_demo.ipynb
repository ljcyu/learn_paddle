```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2019)
N=100

x=np.random.rand(N)
y=10*x**2+x
colors=np.random.rand(N)
area=(20*np.random.rand(N))**2
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
x.shape
x[:5]
area[:3]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

t=np.arange(0.0,2.0,0.01)
s=np.sin(2*np.pi*t)
fig,ax=plt.subplots()

ax.plot(t,s)

ax.set(xlabel='x',ylabel='sin',title='y=sin(x)')
ax.grid()
fig.savefig('sin.png')
plt.show()
```
